```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Read data
iris_data = pd.read_csv('Iris.csv')

#Show Top 5 rows
iris_data.head(5)
```

```python
#Show how many examples we have of each species
iris_data['iris'].value_counts()
```

```python
#Select/filter for Iris-setosa
iris_setosa = iris_data[iris_data['iris']=='Iris-setosa']
iris_virginica = iris_data[iris_data['iris']=='Iris-virginica']
iris_versicolor = iris_data[iris_data['iris']=='Iris-versicolor']
iris_setosa.head(5)

#Basic stats for Iris-setosa
iris_setosa.describe()
```

```python
#More stats for Iris-setosa
iris_setosa['petal length'].min()
```

```python
#Show specific stats for Iris-setosa
iris_setosa.describe(include='all')
```

```python
#Create histogram for the sepal length of the Iris-setosa
plt.hist(iris_setosa['sepal length'], bins=10, histtype='step')
plt.title('Iris Setosa sepal length distribution')
plt.xlabel('sepal length')
plt.ylabel('frequency')
plt.show()
```

```python
#Create scatter plot for all type of Iris looking at the sepal length vs the sepal width
plt.scatter(iris_data['sepal length'], iris_data['sepal width'])
plt.show()
```

```python
#Create boxplots for the petal length of all Iris types
sns.boxplot(x='iris', y='petal length', data=iris_data)
plt.show()
```

```python
#Create violin plots for petal lenght for all Iris types
sns.violinplot(x='iris', y='petal length', data=iris_data)
plt.show()
```

```python
#Create swarm plot
sns.swarmplot(x='iris', y='petal length', data=iris_data)
plt.show()
```

```python
#Create join plot
sns.jointplot(x='petal length', y='petal width', data=iris_data)
plt.show()
```

```python
#Pearson correlation coefficient
np.corrcoef(iris_data['petal length'], iris_data['petal width'])
```
