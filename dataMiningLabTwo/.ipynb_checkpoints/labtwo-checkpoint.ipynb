Importar librerias


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Leer los datos <br>
Los datos están separados por año, por lo que se cargarán
4 conjuntos de datos y posteriormente se trataran como uno sólo.

```python
tsixteen_data = pd.read_csv('2016.csv', encoding='latin-1')
tseventeen_data = pd.read_csv('2017.csv', encoding='latin-1')
teighteen_data = pd.read_csv('2018.csv', encoding='latin-1')
tnineteen_data = pd.read_csv('2019.csv', encoding='latin-1')
```

```python
expo_data = pd.concat([tsixteen_data, tseventeen_data, teighteen_data, tnineteen_data], axis=0)
expo_data.head(5)
expo_data.columns
```

Borrar dataframes viejos para liberar memoria

```python
del tsixteen_data
del tseventeen_data
del teighteen_data
del tnineteen_data
```

Agrupar la información por las variables de interés.
Una vez se agrupa la información por las variables de interés, el resultado queda consolidado en una serie, por lo que es necesario usar el reset_index()

```python
# Agrupar por grupo de producto
expo_modad = expo_data.groupby(['FECH', 'MODAD'])['FOBDOL'].sum().reset_index()
expo_modad.shape
```

```python
# Agrupar por pais
expo_pais = expo_data.groupby(['FECH', 'PAIS'])['FOBDOL'].sum().reset_index()
expo_pais.shape
```

El análisis se hará por periodo, se agrupó por periodo el valor total exportado en
dólares. Para analizar la estabilidad de un tipo de producto analizaremos
la desviación estándar y la frecuencia (para validar que los grupos aparezcan
en todos o la mayoría de los periodos)<br>
 Primero veremos el análisis por país.
<center> <b>Graficas por año por tipo de producto </b>
<br>
<br>

<center> 2016

```python
expo_16 = expo_modad[expo_modad['FECH'] < 1700]
expo_plot = expo_16.pivot(index='FECH', columns='MODAD', values='FOBDOL')
plt.figure(figsize=(10, 6))
plt.plot(expo_plot)
plt.legend(expo_plot.columns, loc="lower center", bbox_to_anchor=(0.5, -0.3), ncol= 5)
plt.show()
```

<center> 2017

```python
expo_17 = expo_modad[(expo_modad['FECH'] > 1700) & (expo_modad['FECH'] < 1800)]
expo_plot = expo_17.pivot(index='FECH', columns='MODAD', values='FOBDOL')
plt.figure(figsize=(10, 6))
plt.plot(expo_plot)
plt.legend(expo_plot.columns, loc="lower center", bbox_to_anchor=(0.5, -0.3), ncol= 5)
plt.show()
```

<center> 2018

```python
expo_18 = expo_modad[(expo_modad['FECH'] > 1800) & (expo_modad['FECH'] < 1900)]
expo_plot = expo_18.pivot(index='FECH', columns='MODAD', values='FOBDOL')
plt.figure(figsize=(10, 6))
plt.plot(expo_plot)
plt.legend(expo_plot.columns, loc="lower center", bbox_to_anchor=(0.5, -0.3), ncol= 5)
plt.show()
```

<center> 2019

```python
expo_19 = expo_modad[expo_modad['FECH'] > 1900]
expo_plot = expo_19.pivot(index='FECH', columns='MODAD', values='FOBDOL')
plt.figure(figsize=(10, 6))
plt.plot(expo_plot)
plt.legend(expo_plot.columns, loc="lower center", bbox_to_anchor=(0.5, -0.3), ncol= 5)
plt.show()
```

Utilizaremos un gráfico tipo violin para analizar a gran escala el
comportamiento de los datos por tipo de producto

```python
plt.figure(figsize=(12, 8))
sns.violinplot('MODAD', 'FOBDOL', data=expo_modad)
plt.show()
```

Dada la gran magnitud de las categorías 198 y 199, lo mejor sería sacar esta categoría
del analisis gráfico para obtener mejores resultados.

Además es claro que la categoría #198 es la menos estable pues se evidencia una desviación estándar muy grande.

```python
categories = expo_modad.MODAD.unique()
for i in categories:
    distplot_data = expo_modad[expo_modad['MODAD']==i]
    plt.figure(figsize=(10, 6))
    plt.title(label=i, loc='center')
    sns.distplot(distplot_data['FOBDOL'])
    plt.show()
```

Revisemos la frecuencia y desviación de los datos por tipo de producto

```python
expo_modad.groupby('MODAD')['FOBDOL'].count()
```

```python
expo_modad.groupby('MODAD')['FOBDOL'].std()
```

Teniendo en cuenta el análisis gráfico y los datos arrojados se podría concluir que
el tipo de producto con mayor estabiliad sería la categoría #199, pues su desviación estándar es baja en comparación con su media y su distribución es la más cercana a la normal (la desviación estándar más baja es la de la categoría #4, pero la media es mucho más baja y sus datos son menos continuos en comparación con la #199)

<center><b> Análisis por pais </b>

```python
expo_pais.head(5)
```

Dada la gran cantidad de paises, reduciremos el análisis a los 10 paises con mayor media de exportación

```python
epgb = expo_pais.groupby('PAIS')['FOBDOL'].mean().reset_index()
```

```python
epgb.sort_values('FOBDOL', ascending=False).head(10)
```

```python
expo_pais_top = expo_pais[expo_pais['PAIS'].isin([249, 580, 493, 827, 239, 573, 589, 105, 215, 245])]
plt.figure(figsize=(12, 8))
sns.violinplot('PAIS', 'FOBDOL', data=expo_pais_top)
plt.show()
```

A partir de esta gráfica es claro que podemos excluir a los paises con código 249, 580 y 493 (que de hecho son los 3 con mayor media) pues sus datos son extremadamente variables

```python
expo_pais_final = expo_pais[expo_pais['PAIS'].isin([827, 239, 573, 589, 105, 215, 245])]
plt.figure(figsize=(12, 8))
sns.violinplot('PAIS', 'FOBDOL', data=expo_pais_final)
plt.show()
```

```python
expo_pais_final.groupby('PAIS')['FOBDOL'].std()
```

Con base a este análisis, podríamos tomar cualquiera de estos paises salvo el 827 como un pais con exportaciones bastante estables, para efectos de análisis tomaremos el pais con código 215, pues es el que tiene una menor desviación estándar

```python
pais_estable = expo_pais[expo_pais['PAIS']==215]
plt.hist(pais_estable['FOBDOL'])
plt.show()
```

El pais con código 215 es China

```python
epgbstd = expo_pais.groupby('PAIS')['FOBDOL'].std().reset_index()
epgbstd.sort_values('FOBDOL', ascending=False).head(10)
```

```python
pais_inestable = expo_pais[expo_pais['PAIS']==249]
plt.hist(pais_inestable['FOBDOL'])
plt.show()
```

Como se puede apreciar en el gráfico, el pais con código 249 (USA) es el pais con desviación mayor desviación estándar, lo cual indica una posible inestabilidad en las exportaciones, lo cual se evidencia en el gráfico (pues hay periodos donde casi no se exporta y periodos donde las exportaciónes son extremadamente altas) <br>
Esto también se puede evidenciar en el primer gráfico de violin de los primeros 10 paises ordenados por media, donde USA fue excluido porque se evidenciaba una gráfica con una kurtosis muy baja.
